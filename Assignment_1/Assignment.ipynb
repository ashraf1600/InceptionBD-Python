# Assignment 1: Temperature Check

```python
temp = float(input("Enter temperature in Celsius: "))
```

```python
if temp < 10 :
    print("It's too cold")

elif 10 <= temp <= 25:
    print("It's cool outside")
else:
    print("It's hot outside")    
```

# Assignment 2: ATM Withdrawal System


```python
balance = float(input("Enter your account balance: "))
```

```python
withdrawal = float(input("Enter amount to withdraw: "))
```

```python
if withdrawal > balance:
    print("Insufficient balance")
else:
    balance -= withdrawal
    print(f"Transaction Successful \n Remaining balance: {balance:.2f}")    
```

# Assignment 3: Student Grade Calculator

```python
marks = float(input("Enter your marks:  "))
```

```python
if 90 <= marks <= 100:
    print("A+")
elif 80 <= marks < 90:  
    print("A")
elif 70 <= marks < 80:
    print("B")
elif 60 <= marks < 70:
    print("C")
else :
    print("Fail")              
```

# Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter ur age:"))
```

```python
if age < 5:
    print("Free")
elif 5 <= age <= 18:
    print(" 50% Discount")
elif age >= 60:
    print(" 30% Discount")  
else:
    print("Full Price")          
```

# Assignment 5: Restaurant Bill with Discount

```python
bill = float(input("Total bill: "))

if bill > 1000:
    discount = bill * 0.1
    bill = bill - discount


print("Final Bill :" , bill)
```

# Assignment 7: Traffic Light System

```python
color = input(" Type a color ( red, yellow, green): ").lower()
if color == "red":
    print("Stop")
elif color == "yellow":
    print("Get Ready")
elif color == "green":
    print("Go")
else:
    print("Invalid color")            
```

# Assignment 8: Leap Year Checker

```python
year = int(input("Enter a year: "))
if year % 400 == 0:
    print("Leap Year")
elif year % 4 ==0 and year % 100 != 0:
    print("Leap Year")

else :
    print("Not a Leap Year")        
```
