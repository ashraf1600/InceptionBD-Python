# 2. Count Word Frequency in Customer Reviews

```python
review = input("Give your review: ")
words = review.lower().split()

unique_words = []   
counts = []        

for word in words:
    if word in unique_words:
        index = unique_words.index(word)  
        counts[index] += 1                 
    else:
        unique_words.append(word)         
        counts.append(1)                 

for i in range(len(unique_words)):
    print(unique_words[i], ":",counts[i])







```

# Password Strength Checker

```python
password = input("Enter Password : ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

for ch in password:
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    elif not ch.isalpha() and not ch.isdigit():
        has_special = True

if has_upper and has_lower and has_digit and has_special:
    print("Strong Password ")
else:
    print("Weak Password ")


```

# 6. Remove Duplicates from Contact List

```python
contact = input("Enter names ")
names = contact.split(",")

unique_names = []

for name in names:
    name = name.strip()
    if name not in unique_names:
        unique_names.append(name)



result = ", ".join(unique_names)

print(result)
```

# 7. Detect Suspicious Words in Text (Moderation)

```python
text = input("Enter text: ")

text_splitted = text.split()
suspicious_words = ["spam", "scam", "fake"]

flag = False

for word in text_splitted:
    if word.lower() in suspicious_words:
        flag = True
        break 

if flag:
    print("Warning: Suspicious content detected!")
else:
    print("Safe: Not suspicious")
```

# 10. Word Reversal for Data Encryption

```python
text = input("Enter a sentence: ")

words = text.split()  
result = ""           

for word in words:
    reversed_word = word[::-1] 
    result = result + reversed_word + " "  

print(result)
```
