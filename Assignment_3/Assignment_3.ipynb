# 1. Inventory Price Calculator


```python
def calculate_inventory_summary(prices):
    if not prices:        
        return (0, 0)

    total_cost = sum(prices)
    average_cost = total_cost / len(prices)

    return (total_cost, average_cost)



prices =list( map ( float , input("Enter item prices separated by spaces: ").strip().split()))
print(calculate_inventory_summary(prices))
```

# 2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    unique_visitors = set(visitor_list)
    return len(unique_visitors)


user_input = input("Enter visitor IDs separated by spaces: ")

visitor_list = list(map(int, user_input.split()))

result = count_unique_visitors(visitor_list)

print("Number of unique visitors:", result)
```

# 3. Product Category Counter


```python
def count_categories(products):
    unique_items = set(products)
    category_count = {}

    for item in unique_items:
        category_count[item] = products.count(item)

    return category_count

user_input = input("Enter product names separated by spaces: ")
products = user_input.split()
result = count_categories(products)
print("Category Counts:", result)
```

# 4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_list = list(filter(lambda x: x > 38, temp_list))
    normal_list = list(filter(lambda x: x <= 38, temp_list))

    return (high_list, normal_list)
user_input = input("Enter temperatures separated by spaces: ")
temp_list = list(map(float, user_input.split()))
result = analyze_temperatures(temp_list)
print("High List (>38째C):", result[0])
print("Normal List (<=38째C):", result[1])
print("Full Output Tuple:", result)
```

# 5. Menu Price Lookup


```python
def get_price(menu, item):
    return menu.get(item, "Item not found")

menu_input = input("Enter menu items with prices (format: item:price item:price ...): ")

menu = {}
for pair in menu_input.split():
    name, price = pair.split(":")
    menu[name] = float(price)
item = input("Enter the item to search: ")
print(get_price(menu, item))
```

# 6. Student Grading System

```python
def calculate_grade(score):
    if score >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 50:
        return "A-"
    else:
        return "F"
def grade_students(student_dict):
    graded = {}
    for student, score in student_dict.items():
        graded[student] = calculate_grade(score)
    return graded

user_input = input("Enter students and scores (format: Name:Score Name:Score ...): ")

students = {}
for pair in user_input.split():
    name, score = pair.split(":")
    students[name] = int(score)
result = grade_students(students)
print("Grades:", result)
```

# 7. Discount Calculator

```python
def apply_discounts(products):
    discounted_prices = []
    for name, price, discount in products:
        discounted_price = price * (1 - discount / 100)
        discounted_prices.append(round(discounted_price, 2))
    return discounted_prices

user_input = input("Enter products (format: Name Price Discount), separated by commas:\n")
products = []
for item in user_input.split(","):
    parts = item.strip().split()
    name = parts[0]
    price = float(parts[1])
    discount = float(parts[2])
    products.append((name, price, discount))
result = apply_discounts(products)
print("Discounted Prices:", result)
```

# 8. Hashtag Generator

```python
def generate_hashtags(sentence):
    words = sentence.split()
    hashtags = ["#" + word.lower() for word in words]
    return hashtags
sentence = input("Enter a sentence: ")

result = generate_hashtags(sentence)
print("Hashtags:", result)
```

# 10. Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
    total = 0
    for item in cart:
        if item in prices:
            total += prices[item]
        else:
            print(f"Warning: {item} not found in price list!")
    return round(total, 2)
cart_input = input("Enter items in cart, separated by spaces: ")
cart = cart_input.split()
price_input = input("Enter prices (format: Item:Price Item:Price ...): ")
prices = {}
for pair in price_input.split():
    name, price = pair.split(":")
    prices[name] = float(price)
total_bill = calculate_total(cart, prices)
print("Total Bill Amount:", total_bill)
```
